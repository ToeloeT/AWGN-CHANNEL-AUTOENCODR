```python
def combine(img1_p,img2_p,save_img):
    img1 = Image.open(img1_p)
    img2 = Image.open(img2_p)
    w = int((img1.size[0] - img2.size[0])/2)
    h = int((img1.size[1] - img2.size[1])/2)
    img1.paste(img2, (w,h))
    img1.save(save_img)
```

```python
for i in snr:
    combine('e2e_blank/snr'+ str(i) +'.png','e2e_result_epch70IQ_alpha0.9-0.8_0.2pi-0.1pi/1t2_snr'+str(i)+'.png','gif_result_epch70IQ_alpha0.9-0.8_0.2pi-0.1pi/frame_'+str(i)+'.png')
```
